# Import Libraries

```python
import IPython
import numpy as np
import pandas as pd
import yfinance as yf
import ipywidgets as widgets
from datetime import datetime
from IPython.display import display
import plotly.graph_objects as go
from plotly.subplots import make_subplots
pd.set_option('display.max_rows', 5000)
import plotly.express as px
```

### Fetching list of companies

```python
company_list = pd.DataFrame(pd.read_excel("NSEcomp.xlsx")) 
company_list['Company Name'] = company_list['Company Name'].str.cat(company_list['Symbol'],sep="-")
```

```python
def calculate_adjusted_prices(df, column):
    adj_column = 'adj_' + column
    df.sort_index(ascending=False, inplace=True)

    price_col = df[column].values
    split_col = df['Splits'].values
    dividend_col = df['Dividends'].values
    adj_price_col = np.zeros(len(df.index))
    adj_price_col[0] = price_col[0]

    for i in range(1, len(price_col)):
        adj_price_col[i] = round((adj_price_col[i - 1] + adj_price_col[i - 1] *
                   (((price_col[i]) -
                     price_col[i - 1] -
                     dividend_col[i - 1]) / price_col[i - 1])), 6)

    df[adj_column] = adj_price_col
    df.sort_index(ascending=True, inplace=True)

    return df
```

```python
def fetch_data():
    global df
    df = pd.DataFrame(yf.download(Company.value[Company.value.index('-')+1:] + '.NS', period=Duration.value, interval=Interval.value))
    df = df.dropna()
    msft = yf.Ticker(Company.value[Company.value.index('-')+1:] + '.NS')
    global dr
    dr = pd.DataFrame(msft.dividends)
    global dp
    dp = pd.DataFrame(msft.splits)
    df["Dividends"]=dr["Dividends"]
    df = df.fillna(0)
    df["Splits"]=dp["Stock Splits"]
    df = df.fillna(1)
    df=df.reset_index()
    calculate_adjusted_prices(df, 'Open')
    calculate_adjusted_prices(df, 'High')
    calculate_adjusted_prices(df, 'Low')
    calculate_adjusted_prices(df, 'Close')
```

```python
companies = company_list["Company Name"].tolist()
Interval = widgets.SelectionSlider(
    options=['1m', '2m', '5m', '15m', '30m', '60m', '90m', '1h', '1d', '5d', '1wk', '1mo', '3mo'],
    value='1d',
    description='Interval',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True
)

Company = widgets.Combobox(
    placeholder='Company',
    options=companies,
    description='Company:',
    ensure_option=True,
    disabled=False
)
 
Duration = widgets.Text(
    value='1y',
    placeholder='Type something',
    continuous_update=False,
    description='Duration:',
    disabled=False
)

def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        IPython.display.clear_output(wait=False)
        display(Company)
        display(Duration)
        display(Interval)
        fetch_data()
        print(df.head(10))
        maxi = df['High'].max()
        mini = df['High'].min()
        fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                open=df['adj_Open'],
                high=df['adj_High'],
                low=df['adj_Low'],
                close=df['adj_Close'])])
        fig.update_layout( title='Historical Data of Stock Prices', 
                          yaxis_title=Company.value+" Stock Price ------->",
                          xaxis_title=" Time ------->",)
        fig.layout.plot_bgcolor = '#d1d1cd'
        fig.layout.paper_bgcolor = '#d1d1cd'
        for dt in df[df['Dividends']>0]['Date'].tolist():
            yi = df[df['Date']==dt]['adj_High'].tolist()[0]
            temp = 0
            if yi > (maxi+mini)/2:
                temp = -(maxi-mini)/50
            else:
                temp = (maxi-mini)/50
            fig.add_annotation(x=dt, 
                       y=yi,
                       xref="x",
                       yref="y",
                       text="D",
                       showarrow=True,
                       bgcolor="#ff7f0e", 
#                        arrowside = "end+start",
                       hovertext = str(dt)[0:10] + " Dividend = " + str(sum(df[df['Date'] == dt]['Dividends'].tolist())),
                       opacity=0.8)

        for dt1 in df[df['Splits']!=1.0]['Date'].tolist():
            yi = df[df['Date']==dt1]['adj_High'].tolist()[0]
            temp = 0
            if yi > (maxi+mini)/2:
                temp = -(maxi-mini)/50
            else:
                temp = (maxi-mini)/50
            fig.add_annotation(x=dt, 
                       y=yi + 10,
                       xref="x",
                       yref="y",
                       text="S",
                       showarrow=True,
                       hovertext = str(dt)[0:10] + " Stock Split = " + str(sum(df[df['Date'] == dt1]['Splits'].tolist())),
                       bgcolor="#ff7f0e",
                       opacity=0.8)
        fig.show()


Company.observe(on_change)
Duration.observe(on_change)
Interval.observe(on_change)
display(Company)
display(Duration)
display(Interval)
```

```python
EMA_12 = df['Close'].ewm(span = 12).mean()
EMA_26 = df['Close'].ewm(span = 26).mean()
df['MACD'] = EMA_12 - EMA_26

fig = go.Figure(data=[go.Candlestick(x=df.Date,
                                     open=df.Open, 
                                     high=df.High,
                                     low=df.Low,
                                     close=df.Close), 
                      go.Scatter(x=df.Date, y=df.MACD, line=dict(color='green', width=2))])

fig.show()
```
